```python
import pandas as pd
df_temporal = pd.read_csv("./datasets/popularidad/temporal.csv")
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    """
    display dataframes in row
    """
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
def exhaustiveDescribe(p_dataframe) : 
    """
    @parameter = Dataframe
    Display info, sum of nan et describe from a data frame in row 
    using "display_side_by_side" function from IPython.display import display_html
    """
    display_side_by_side(
        p_dataframe.head(6),
        pd.DataFrame(p_dataframe.info()),
        pd.DataFrame(p_dataframe.isna().sum(), columns=["Number of NaN"]),
        pd.DataFrame(p_dataframe.describe())
                        )
```

```python
exhaustiveDescribe(df_temporal)
```
